CountVectorizer: counts how many times each word appears and puts them into a vector

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
```

```python
# discovers whic words are present in dataset
dvec = vectorizer.fit_transform(d['CONTENT'])
```

```python
# gives bag of words matrix for phrases in count
dvec
```

```python
print(d['CONTENT'][0])
#analyze(d['CONTENT'][0])
```

```python
# to find out which word the dataset found after vectorizing
vectorizer.get_feature_names()
```

```python
# to shufffle dataset with fraction 100%
data_shuffle = d.sample(frac=1)
```

```python
data_shuffle.head()
```

```python
d_train = data_shuffle[:300]
d_test = data_shuffle[300:]
d_train_att = vectorizer.fit_transform(d_train['CONTENT'])  #fit bag of words on training set(learn the words and produce amatrix)
d_test_att = vectorizer.transform(d_test['CONTENT'])       # reuse on testing set
d_train_label = d_train['CLASS']
d_test_label = d_test['CLASS']
```

```python
d_train_att
```

```python
d_test_att
```

Using Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=80)   #converting dataset into 80 different trees
```

```python
clf.fit(d_train_att, d_train_label)   #fitting trainig set
```

```python
clf.score(d_test_att, d_test_label)   # calculating score performance on testing set
```

```python
from sklearn.metrics import confusion_matrix
pred_labels = clf.predict(d_test_att)
confusion_matrix(d_test_label, pred_labels)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf,d_train_att, d_train_label, cv=5)

print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(),scores.std()*2))
```

```python
#shuffling entire datset and separating coments and class answers
dshuff = d.sample(frac=1)
d_content = dshuff['CONTENT']
d_label = dshuff['CLASS']
```

'''Pipeline is really convenient and will bring together two or more steps so that all the steps are 
treated as one. So, we will build a pipeline with the bag of words, and then use countVectorizer
followed by the random forest classifier. Then we will print the pipeline.
'''

```python
#set up a pipeline
from sklearn.pipeline import Pipeline, make_pipeline
pipeline = Pipeline([
    ('bag-of-words',CountVectorizer()),
    ('random forest',RandomForestClassifier()),
])
pipeline
```

```python
# say combining CountVectorizer and RandomForestClassifer() into one
make_pipeline(CountVectorizer(),RandomForestClassifier())
```

```python
#first perform fit then transform with CountVectorizer followed by fit of randomForest classifier
pipeline.fit(d_content[:1500],d_label[:1500])
```

```python
pipeline.score(d_content[1500:], d_label[1500:])
```

```python
pipeline.predict(["What a project Aditya!"])
```

```python
pipeline.predict(["Please vit guys subscribe to my youtube channel"])
```

```python
pipeline.predict(["have a look to my channel"])
```

```python
#cross validation
scores = cross_val_score(pipeline, d_content, d_label,cv=5)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std()* 2))
```
